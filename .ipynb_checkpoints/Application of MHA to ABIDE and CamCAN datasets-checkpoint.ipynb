## Python implementation of Monti-Hyvarinen Analysis (MHA) algorithm

#### The MHA model is an extension of traditional factor analytic and PCA models. In particular, MHA enjoys the following benefits over traditional methods such as PCA and factor analysis
 * MHA is identifiable - this means that the latent variables which generated the data can be recovered (up to some tolerable indeterminacies)
 * MHA is built to facilitate the interpretation of results 


#### For further details see:
 * Monti & Hyvarinen, "A Unified Probabilistic Model for Learning Latent Factors and Their Connectivities from High-Dimensional Data", UAI 2018
 * Monti et al., "Interpretable brain age prediction using linear latent variable models of functional connectivity", under review 2019

```python
import numpy as np 
import os 
import pandas as pd 
import pylab as plt 
from sklearn.preprocessing import scale 

%matplotlib inline

from MHA import MHA 
```

```python
# we begin with the ABIDE data
os.chdir('/Users/ricardo/Documents/Projects/Score matching GGM/ABIDE application/subjectData_recent/rois_aal/UM_1/')

files = os.listdir( os.getcwd() ) 
files = [f for f in files if '.csv' in f ]

SubDat = [ scale( np.array( pd.read_csv(f) ) ) for f in files  ]
Shat = [ np.cov( S.T ) for S in SubDat if np.isnan(S).sum()==0 ]

print( 'Collected data for ' + str( len( Shat ) ) + ' subjects' )

# also load in the 
ROIcoord = np.loadtxt('/Users/ricardo/Documents/Projects/Score matching GGM/ABIDE application/AAL_coord_mm.txt')
```

```python
# run the MHA algorithm
k=5
#res = nonNegativeCovFactor_LagrangeMult(Shat=Shat, k=k, diagG=True, lagParam=.5, tol=.01 )
res = MHA( Shat=Shat, k=k )
res.fit( lagParam =.5 , tol=.01, alphaArmijo=1 )
res.iter
```

```python
# visualize some estimated networks 
for k in range(5):
    res.plot( ROIcoord, clusterID=k )
```
